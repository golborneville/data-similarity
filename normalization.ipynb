```python
def z_transform(data):
    # x' = (x - mean(x))/(std(x))
    data_len = len(data)
    x_sc = []
    for i in range(len(data)):
        x_sc.append((data[i] - data.mean())/data.std())
    return x_sc
```

```python
#직접 함수구현
print(z_transform(data.iloc[0]))
```

# scale min max (normalization)

```python
def scaleminmax(data): #하나의 종목에 대해서만 정규화 = normalization (0,1)
    # x' = (x - min(x))/(max(x) - min(x))
    data_len = len(data)
    x_sc = []
    for i in range(len(data)):
        x_sc.append((data[i] - min(data))/(max(data) - min(data)))
    return x_sc
```

```python
print(scaleminmax(data.iloc[0])) # 직접 구현 -> 얘 scale 이 아니라 minmaxsale이엇어; 출력은 list여
```

# list to dataframe

```python
df_index= ['A005930','A000660','A068270','A005380','A005490', 'A035420','A051910',
           'A105560','A012330','A055550','A015760','A051900','A032830','A017670','A034730',
           'A096770','A090430','A006400','A033780','A066570','A000270',
           'A003550','A010950','A086790','A000810','A011170','A009150','A000030']
```

```python
def list_to_dataf(data):
    dlen = len(data)
    dlist = []
    for i in range(dlen):
        dlist.append(z_transform(data.iloc[i]))
    ldf = pd.DataFrame(dlist, index = df_index)
    return ldf
```

```python
result = list_to_dataf(data) #list ->dataframe (type)
```
